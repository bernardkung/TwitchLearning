### Tokenization

tokenize messages, create a token count

```python
log['token'] = log.apply(lambda row: nltk.word_tokenize(row['message']), axis = 1)
```

```python
log['token_count'] = log['token'].apply(lambda x: len(x))
```

```python
log_agg["avg_token_count"] = log[["timestamp", "token_count"]].groupby("timestamp").mean()
log_agg.head()
```

```python
log_agg = log_agg.reset_index()
log_agg = log_agg.rename(columns = {'message':'num_message'})
log_agg.head()
```

```python
message = log['message'].str.cat(sep=' ')
message_tokens = nltk.word_tokenize(message)
```

```python
def lexical_diversity(text):
    return len(set(text)) / len(text) 

def percentage(count, total):
     return 100 * count / total
```

```python
lexical_diversity(message_tokens)
```

```python
message_freq = nltk.FreqDist(message_tokens)
message_freq.plot(20, cumulative=False)
```

Is there a way to determine the most commonly used emote?


```python
message_unique_tokens =  set(message_tokens)
```

```python
emotes = pd.read_json('https://twitchemotes.com/api_cache/v3/global.json').transpose()
```

```python
emotes.head()
```

```python
message_emotes = [t for t in message_unique_tokens if t in emotes['code']]
message_emotes
```

```python
msg_emote_count = [message_tokens.count(tkn) for tkn in message_emotes]
msg_dict = {"emote": message_emotes, "count": msg_emote_count}
msg_df = pd.DataFrame(msg_dict)
msg_df.head()
```

```python
msg_df= msg_df.sort_values('count', ascending=False).reset_index(drop=True)
```

```python
msg_df.head(15)
```

```python
plt.figure(figsize=(8, 8))
sns.barplot(data=msg_df.head(20), x='count', y='emote')
```

Seems to be a problem here, the chat log goes for 10 hours, starting at 12:41AM and going until 10:30AM. I might need something more sophisticated than arbitrarily determining chat intensity.
