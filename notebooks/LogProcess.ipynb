Define and demonstrate a custom _logProcess()_ function for reading in Hexchat .log files, clean and format into a pandas DataFrame, and then exporting the DataFrame as a .json file for further analysis.

```python
import pandas as pd
import os
```

```python
def logProcess(input_name):
    input_path = os.path.join('..\\Logs\\', input_name)
    with open(input_path, encoding = 'latin1') as f:
        inlog = f.readlines()
    input_log = pd.DataFrame(inlog[4:-4])
    input_log[['key','message']] = pd.DataFrame(input_log[0].str.split('\t').tolist(), 
                                                columns = ['key','message'])
    input_log[['month','day','time','username']] = pd.DataFrame(input_log['key'].str.split().tolist()).iloc[:,0:4]
    input_log[['username']] = pd.DataFrame(input_log['username'].str.lstrip('<').str.rstrip('>'))
    input_log[['message']] = pd.DataFrame(input_log['message'].str.rstrip('\n'))
    input_log['timestamp'] = pd.to_datetime(input_log.month + " " + input_log.day.astype(str) + ", " + "2018"+ 
                                  " " + input_log.time.astype(str))
    export_log = input_log[['timestamp','username','message']]
    export_name = input_name.split('.')[0] + '.txt'
    export_path = os.path.join('..\\Data\\', export_name)
    export_json = pd.DataFrame.to_json(export_log, orient='split')
    with open(export_path, 'w') as f:
        f.write(export_json)
    print(export_name, "successfully processed!")
```

```python
log_files = os.listdir("..\\logs\\")
```

```python
logProcess(log_files[9])
```

```python
[logProcess(item) for item in log_files]
```
