```python
input_name = 'owl3.log'
```

```python
input_path = os.path.join('..\\Logs\\', input_name)
```

```python
with open(input_path, encoding = 'latin1') as f:
    inlog = f.readlines()
```

* inlog is a list read directly from the log files.

* The first four and last four lines are the beginning and end of log announcements, respectively, and should be removed. 

* Finally, each row should be split.

    * Split once on '\t' to remove messages
    * Split on whitespace to get month, date, time, and username
    * lstrip and rstrip to remove the brackets from usernames
    * rstrip to remove the newline '\n' at the end of messages

```python
input_log = pd.DataFrame(inlog[4:-4])
```

```python
input_log[['key','message']] = pd.DataFrame(input_log[0].str.split('\t').tolist(), 
                                                 columns = ['key','message'])
```

```python
input_log[['month','day','time','username']] = pd.DataFrame(input_log['key'].str.split().tolist(), 
                                                 columns = ['month','day','time','username'])
```

```python
input_log[['username']] = pd.DataFrame(input_log['username'].str.lstrip('<').str.rstrip('>'))
```

```python
input_log[['message']] = pd.DataFrame(input_log['message'].str.rstrip('\n'))
```

```python
input_log.head()
```

```python
export_log = input_log[['month','day','time','username','message']]
```

```python
export_name = input_name.split('.')[0] + '.txt'
```

```python
export_name
```

```python
export_path = os.path.join('..\\Data\\', export_name)
```

```python
export_log.head()
```

```python
export_json = pd.DataFrame.to_json(export_log, orient='split')
with open(export_path, 'w') as f:
    f.write(export_json)
```
